```python
# import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Creaate the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()  2nd method
fig=plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
#To create a quick graph using the object-oriented interface method, use the following two lines:
ax=plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#Create the plot and add a label ofr the legend.
plt.plot(x_axis, y_axis, label= "Boston")

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

#Set the y limit between 0 and 45
plt.ylim(0, 45)

#Create a title.
plt.title("PyBer fare by Month")

# add a legend
plt.legend()
```

```python
#Create the plot
plt.plot(x_axis, y_axis, marker="*", color = "blue", linewidth=2, label="Boston")

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

#Set the y limit to between 0 adn 45
plt.ylim(0, 45)

#Create the title
plt.title("PyBer Fare by Month")

#add a grid.
plt.grid()

#add a legend
plt.legend()
```

## Bubble Charts


```python
# Create a bubble chart, note the bubbles are in relation to the values in the data series y.
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#make the bubbles bigger.
y_axis_larger = []
for data in y_axis:
        y_axis_larger.append(data*3)
        
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#introduce a list comprehension - same result as the for loop above.
# new_list = [exression for item in list if conditional]
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
#object oriented scatter plot
fig, ax=plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
#object oriented bubble scatter plot
fig, ax=plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

### Skill Drill
Change the color of the markers to sky blue.

## Pie Charts

```python
#create a pie chart using Matplotlib.
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values=(0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode = explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
#assign 12 colors one for each month, 
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.subplots(figsize=(8,8))
plt.pie(y_axis, 
        explode=explode_values,
       colors = colors,
       labels= x_axis,
       autopct='%.1f%%')

plt.show()

```

```python
#create a pie chart using object oriented interface
fig, ax=plt.subplots()
ax.pie(y_axis, labels= x_axis)

plt.show
```
